```python
import pygplates
import glob
import os

list_of_tween_files = glob.glob('./tween_feature_collections/*.gpmlz')

out_dir = './tween_feature_collections_gpml/'

for tween_file in list_of_tween_files:
    
    print tween_file
    filename = os.path.splitext(os.path.split(tween_file)[1])
    pygplates.FeatureCollection(tween_file).write('%s/%s.gpml' % (out_dir,filename[0]))
```
